# Liens possibles entre les databases

'customer_id' est pr√©sent dans *olist_customers_dataset.csv* et *olist_orders_dataset.csv*, donc on doit pouvoir relier les **consommateurs** et les **commandes**.

'order_id' est pr√©sent dans *olist_order_items_dataset.csv*, *olist_order_payments_dataset.csv*, *olist_order_reviews_dataset.csv* et *olist_orders_dataset.csv* donc on doit pouvoir relier les **produits achet√©s**, les **moyens de paiements**, les **avis sur les commandes** et les **commandes**.

'seller_id' est pr√©sent dans *olist_order_items_dataset.csv* et *olist_sellers_dataset.csv* donc on doit pouvoir relier les les **produits achet√©s** et les **vendeurs**.

'product_id' est pr√©sent dans *olist_order_items_dataset.csv*, *olist_products_dataset.csv* et *product_category_name_translation.csv* donc on doit pouvoir relier les **produits achet√©s**, les **produits disponibles** et les **cat√©gories de produit en anglais**.

Le lien avec la database restante traitant la **g√©olocalisation** peut se faire via 'customer_city' dans *olist_customers_dataset.csv* et 'geolocation_city' dans *olist_geolocation_dataset.csv* (mais il doit y avoir plein de geolocalisations dans une m√™me ville, ces donn√©es ne seront peut-√™tre pas exploitables).

Pour *olist_order_reviews_dataset.csv*, il faudra s√ªrement utiliser 'order_id' (quitte √† fusionner des lignes si besoin pour en faire une cl√© unique, puisqu'elle est d√©j√† utilis√©e dans les fusions pr√©c√©dentes). Nous aurons alors les **avis sur les commandes** ajout√©s √† notre base de donn√©es.


```python
#V√©rifions si les variables sont uniques et donc exploitables pour rattacher les databases
for i in range(9):
    data = pd.read_csv(names[i])
    #On affiche le nom du fichier
    print(names[i])
    #On affiche le nombre d'√©l√©ments moins le nombre d'√©l√©ments DISTINCTS, si on atteint 0 on a une cl√© primaire
    print(data.count()-data.nunique())
    #Id√©alement il faut au moins une cl√© primaire pour chaque fusion (dans une des deux bases concern√©es)
```

```python
#On va charger et enregistrer toutes les databases utiles pour les fusionner
customers = pd.read_csv('olist_customers_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
order_reviews = pd.read_csv('olist_order_reviews_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
sellers = pd.read_csv('olist_sellers_dataset.csv')
categories = pd.read_csv('product_category_name_translation.csv')
#On laisse de c√¥t√© geolocations et order_payments qui ne nous seront pas utiles

#On fusionne les bases de donn√©es en une seule
data = customers.merge(orders, how='outer', on='customer_id', validate="1:1")
data = data.merge(order_items, how='right', on='order_id', validate="1:m")
data = data.merge(sellers, how='left', on='seller_id')
data = data.merge(products, how='left', on='product_id')
data = data.merge(categories, how='left', on='product_category_name')
#Pour la partie "reviews" il faut d'abord cr√©er une cl√© primaire en agglom√©rant les reviews d'une m√™me commande
#puisque 'order_id' n'est plus une cl√© primaire dans notre database √† cause des merge successifs
reviews = order_reviews.groupby('order_id').agg(list)
data = data.merge(reviews, how='left', on='order_id')

#On affiche la database nouvellement cr√©√©e
print(data.shape)
data.head(10)
```

```python
#On va jeter un oeil aux doublons de "customer_id" pour voir si la duplication des lignes lors de la fusion a bien march√©
colonnes = ['customer_id', 'order_id', 'product_id', 'seller_id', 'price', 'product_category_name_english']
doublons = data[data.duplicated(subset=['customer_id'], keep=False)][colonnes]
print(doublons.shape)
print(doublons.head(10))
```

On va chercher √† appliquer la m√©thode RFM, qui s√©pare les clients selon la R√©cence de leurs commandes, la Fr√©quence de celles-ci ainsi que leurs Montants. Il nous faut donc d√©terminer ces variables

```python
#On trie selon la date de la derni√®re commande
data = data.sort_values(by=['order_purchase_timestamp'], ascending=False)
data['order_purchase_timestamp'].head()
```

```python
#Il faudra formater les dates qui sont au format string
def format_date(date):
    if type(date) == str:
        return dt.datetime.strptime(date, '20%y-%m-%d %H:%M:%S')
    else:
        return 'Nan'

#On r√©cup√®re la derni√®re date de commande dans nos donn√©es
today = data['order_purchase_timestamp'].iloc[0]
#On la formate
aujourdhui = format_date(today)
```

```python
#On formate les dates de derni√®re commande
data['order_purchase_datetime'] = data['order_purchase_timestamp'].apply(lambda x : format_date(x))
```

```python
#avant d'en d√©duire la r√©cence par soustraction √† la date la plus tardive
def recence(date, today):
    if date != 'Nan':
        return today - date
data["Recence"] = data['order_purchase_datetime'].apply(lambda x : recence(x, aujourdhui))
```

```python
#On affiche un extrait (d√©j√† tri√©) pour voir cette nouvelle variable
data["Recence"].head(10)
```

```python
#Boxplot pour voir la r√©partition de ces dates
plt.boxplot(data["Recence"])
plt.show()
```

```python
#On cr√©e une version qui ne conserve que le nombre de jours depuis la derni√®re commande
data["Recence_days"] = data["Recence"].astype('timedelta64[D]')
#On affiche sa r√©partition, en boxplot et en densit√©
plt.boxplot(data["Recence_days"].dropna())
plt.show()
plt.hist(data["Recence_days"], bins=50, color='steelblue', density=True, edgecolor='none')
plt.show()
```

```python
#On r√©cup√®re la r√©cence de la derni√®re commande
recence_last = data.groupby('customer_unique_id')['Recence_days'].agg('min')
recence_last.rename('Recence_derniere_cmd', inplace=True)
#On fusionne avec notre database
data = data.merge(recence_last, how='left', on='customer_unique_id')
```

```python
#On affiche sa r√©partition, en boxplot et en densit√©
plt.boxplot(data["Recence_derniere_cmd"].dropna())
plt.show()
plt.hist(data["Recence_derniere_cmd"], bins=50, color='steelblue', density=True, edgecolor='none')
plt.show()
```

```python
#On va r√©cup√©rer le nombre de commandes pass√©es par chaque consommateur
nombre_commandes = data.groupby(by=['customer_unique_id'])['order_id'].nunique()
nombre_commandes.describe()
```

```python
#On affiche la densit√©
plt.hist(nombre_commandes.dropna(),
            bins=50, color='steelblue', density=True, edgecolor='none')
plt.show()
```

```python
#On affiche la r√©partition des diff√©rentes valeurs (qui semblent √™tre peu nombreuses)
nombre_commandes.value_counts()
```

**Seulement 3% d'utilisateurs dans notre base de donn√©es ont pass√© plusieurs commandes, comme annonc√© par Olist.**

```python
#On va ajouter cette variable √† notre base de donn√©es
nombre_commandes.rename('Nombre_commandes', inplace=True)
data = data.merge(nombre_commandes, how='left', on='customer_unique_id', validate="m:1" )
```

```python
#On va cr√©er une variable bool√©enne qui distingue simplement les commandes uniques des commandes multiples
data["Plusieurs_commandes"] = data['Nombre_commandes'].apply(lambda x : x > 1)
data.head()
```

```python
#On r√©cup√®re le montant total des commandes de chaque consommateur
montant_total = data.groupby(by=['customer_unique_id'])['price'].sum()
montant_total.rename('Montant_total', inplace=True)
data = data.merge(montant_total, how='left', on='customer_unique_id', validate="m:1" )
```

```python
#On affiche la r√©partition
plt.hist(montant_total.dropna(),
            bins=50, color='steelblue', density=True, edgecolor='none')
plt.show()
```

```python
#On affiche la r√©aprtition jusqu'√† un maximum pour plus de lisibilit√©
plt.hist(montant_total.dropna(), range=(0,1000),
            bins=50, color='steelblue', density=True, edgecolor='none')
plt.show()
```

```python
#On r√©cup√®re le montant de la derni√®re commande, m√™me si peu d'√©l√©ments devraient diff√©rer par rapport √† la variable pr√©c√©dente
#On somme les prix des articles pour chaque commande
montant_last = data.groupby(by=['order_id'])['price'].sum()
montant_last.rename('Montant_dernier', inplace=True)
#On fusionne cette colonne avec les id des commandes, des clients ainsi que la date des commandes
montant_dernier = data[['customer_unique_id', 'order_id', 'order_purchase_timestamp']]\
        .merge(montant_last, on='order_id', how='left')
#On peut alors ordonner par date pour ne conserver que la commande la plus r√©cente pour chaque consommateur
montant_dernier.sort_values(by=['order_purchase_timestamp'], ascending=False, inplace=True)
montant_dernier.drop_duplicates(subset=['customer_unique_id'], keep='first', inplace=True)
```

```python
#On affiche la r√©partition
plt.hist(montant_last.dropna(),
            bins=50, color='steelblue', density=True, edgecolor='none')
plt.show()
```

```python
#On affiche la r√©aprtition jusqu'√† un maximum pour plus de lisibilit√©
plt.hist(montant_last.dropna(), range=(0,1000),
            bins=50, color='steelblue', density=True, edgecolor='none')
plt.show()
```

```python
#On fusionne cette variable avec notre base de donn√©es
data = data.merge(montant_dernier[['customer_unique_id', 'Montant_dernier']], how='left',
                  on='customer_unique_id', validate="m:1" )
```

```python
#On r√©cup√®re le montant moyen des articles command√©s par chaque consommateur
prix_moyen = data.groupby(by=['customer_unique_id'])['price'].mean()
prix_moyen.rename('Prix_moyen_articles', inplace=True)
data = data.merge(prix_moyen, how='left', on='customer_unique_id', validate="m:1" )
```

```python
#On affiche la r√©partition
plt.hist(prix_moyen.dropna(),
            bins=50, color='steelblue', density=True, edgecolor='none')
plt.show()
```

```python
#On va g√©n√©rer le review_score moyen des utilisateurs en ayant donn√©
data['review_score'].value_counts()
```

```python
#On vient cr√©er la moyenne des notes donn√©es lors des diff√©rents retours d'un m√™me consommateur
review_scores = data[['customer_unique_id', 'review_score']]
mean_review_score = review_scores.groupby('customer_unique_id').agg(sum)
mean_review_score.head()
```

```python
#On va afficher la forme et les statistiques du pd.series cr√©√© pour s'assurer qu'il n'y a aucun probl√®me
print(mean_review_score.shape)
mean_reviews = mean_review_score['review_score'].apply(np.mean)
mean_reviews.describe()
```

```python
#On affiche la r√©partition de la note moyenne donn√©e par chaque consommateur
plt.hist(mean_reviews.dropna(), bins=50, color='steelblue', density=True, edgecolor='none')
plt.show()
```

```python
#On fusionne cette nouvelle variable avec notre database
mean_reviews.rename("Mean_review_score", inplace=True)
data = data.merge(mean_reviews, on='customer_unique_id', how='left')
```
